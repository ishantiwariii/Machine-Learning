## Regression analysis 

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
df = pd.read_csv(r"D:\Machine Learning\regression analysis\house_price_data.csv")
df
```

```python
x = df[['Age_of_House', 'Square_Footage']]
y = df['Price_1k']
```

```python
X = sm.add_constant(x)
```

```python
model = sm.OLS(y,X)
result = model.fit()
```

```python
print(result.summary())
```

```python
residuals = result.resid


plt.scatter(result.fittedvalues, residuals)
plt.axhline(0)
plt.show()
```
