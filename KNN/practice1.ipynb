### KNN - K nearest neighbours 

`The core philosophy of KNN can be summarized as: "Tell me who your neighbors are, and I will tell you who you are." It is a non-parametric, lazy learning algorithm.`

- Non-parametric: It makes no underlying assumptions about the distribution of the data.
- Lazy learning: It doesn't build a model during the training phase. Instead, it memorizes the training dataset and does all the heavy lifting (computation) at the time of prediction.

`How it works:`

- Choose the number of neighbors, $K$.
- When a new data point arrives, calculate its distance (typically Euclidean distance, $$d(p, q) = \sqrt{\sum_{i=1}^{n} (q_i - p_i)^2}$$) from all other points in the dataset.
- Identify the $K$ points closest to the new data point.
- For Classification: Assign the new data point to the class that has the majority vote among those $K$ neighbors.
- For Regression: Assign the average value of those $K$ neighbors to the new data point.

### here is code example for knn

```python
import pandas as pd
import numpy as np 
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"D:\Machine Learning\KNN\data.csv")
df.head()
```

```python
df.drop(columns=['id','Unnamed: 32'],inplace=True)
df.head(5)
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:,1:], df.iloc[:,0],test_size=0.2, random_state=2)
```

```python
X_train.head()
```

```python
scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score

y_pred = knn.predict(X_test)

accuracy_score(y_test, y_pred)
```

```python
scores = []

for i in range(1,16):
    
    knn = KNeighborsClassifier(n_neighbors=i)
    
    knn.fit(X_train,y_train)
    
    y_pred = knn.predict(X_test)

    scores.append(accuracy_score(y_test, y_pred))
```

```python
plt.plot(range(1,16),scores)
```
